```python
import pandas as pd
```

```python
data = pd.read_csv("Data.csv", sep = ';')
```

```python
data
```

```python
data['Position'].value_counts()
```

```python
data['Goals'] = data['Goals'].replace("-","0",regex=True)
data['Assists'] = data['Assists'].replace("-","0",regex=True)
data['Yel'] = data['Yel'].replace("-","0",regex=True)
data['Red'] = data['Red'].replace("-","0",regex=True)
data['SpG'] = data['SpG'].replace("-","0",regex=True)
data['PS%'] = data['PS%'].replace("-","0",regex=True)
data['MotM'] = data['MotM'].replace("-","0",regex=True)
data['AerialsWon'] = data['AerialsWon'].replace("-","0",regex=True)
```

```python
data
```

```python
data.drop(data.tail(1).index,inplace=True) # drop last n rows
```

```python
data = data.drop(columns=['Unnamed: 16'])
```

```python
data
```

```python
data['CM'] = data['CM'].astype(int)
data['Age'] = data['Age'].astype(int)
data['KG'] = data['KG'].astype(int)
data['Apps'] = data['Apps'].astype(int)
data['Mins'] = data['Mins'].astype(int)
```

```python
data
```

```python
data.rename(columns={'CM': 'Height(Cm)',
                     'KG': 'Weight(Kg)',
                     'Apps': 'Appearances',
                     'Yel': 'Yellow',
                     'SpG': 'Shots per game',
                     'PS%': 'Pass success (%)',
                     'AerialsWon': 'Aerials Won'
                    }, inplace = True)
```

```python
data
```

```python
data.to_csv('Dados Formatados.csv', sep = ';')
```
